```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns_to_encode = ['Возрастные группы', 'Занятость', 'Количество человек в семье', 
                     'Образование', 'Пол', 'Семейное положение', 'Уровень дохода',
                     'Федеральный округ (БМ и СПб отдельно)', 'Численность населенного пункта']

for column in columns_to_encode:
    df_resp[column] = le.fit_transform(df_resp[column]) + 1
    
df_resp
```

```python
columns_to_drop = ['SubjectID', 'Population', 'dt']
df_resp = df_resp.drop(columns=columns_to_drop)

df_resp
```

```python
from sklearn.linear_model import LogisticRegression

df_ml = df_resp

# Определяем признаки (X) и целевую переменную (y)
X = df_ml.drop(columns=['Занятость'])
y = df_ml['Занятость']

# Создаем и обучаем модель логистической регрессии
model = LogisticRegression()
model.fit(X, y)

predicted_values = model.predict(X)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, predicted_values)
print("Точность оценки параметра y:", accuracy)
```

```python
import numpy as np
from scipy.stats import binom
from scipy.optimize import minimize

# Генерим выборку из биномиального распределения
n = 1000  # количество испытаний
p = 0.5  # вероятность успеха
sample = np.random.binomial(n, p, size=100)

# Определяем функцию правдоподобия
def likelihood(params):
    p = params[0]
    return -np.sum(binom.logpmf(sample, n, p))

# Поиск параметра, максимизирующего функцию правдоподобия
result = minimize(likelihood, [0.5], method='Nelder-Mead')
estimated_p = result.x[0]

print("Оценка параметра p:", estimated_p)
```

```python
import numpy as np
from scipy.optimize import minimize

sample = df_activity['dur']

# Определение функции правдоподобия
def likelihood(params):
    mu, sigma = params
    log_likelihood = -np.sum(np.log(sigma) + 0.5*np.log(2*np.pi) + ((sample - mu)**2) / (2*sigma**2))
    return -log_likelihood

# Начальные значения параметров
initial_params = [50,20]
# initial_params = [df_activity['dur'].mean(), df_activity['dur'].var()]

# Минимизация отрицательного логарифма функции правдоподобия
result = minimize(likelihood, initial_params)

# Оценки параметров методом максимального правдоподобия
mu_mle, sigma_mle = result.x

print(f"Реальное среднее: {df_activity['dur'].mean()}")
print(f"Реальная дисперсия: {df_activity['dur'].var()}")

print(f"Оценка среднего значения: {mu_mle}")
print(f"Оценка дисперсии: {sigma_mle}")
```
